# **1. How do sales change over time (daily , monthly, yearly) ? Arethere any sesonal patterns or significant trends??**

```python
df['Sale_Date']=pd.to_datetime(df['Sale_Date'])
df=df.sort_values(by='Sale_Date')

df['Month']=df['Sale_Date'].dt.to_period("M")
df_monthly=df.groupby("Month").agg({"Sales_Amount" : "sum"}).reset_index()

df_monthly["Month"]=df_monthly["Month"].astype(str)
plt.figure(figsize=(10,5))

sns.lineplot(data=df_monthly, x="Month", y="Sales_Amount",markers="o", linewidth=2)
plt.xticks(rotation=45)

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales Amount")
plt.grid(True)
plt.show()
```

Analysis of Sales Trends Based on the Graph

1. Monthly Sales Trends
Sales Fluctuate Over Time: There are noticeable ups and downs, indicating that sales do not remain constant every month. Some months experience higher sales, while others dip.

Seasonal Patterns: If the peaks occur around the same months each year, this could indicate seasonal demand (e.g., higher sales during holiday seasons, festivals, or end-of-quarter periods).

Significant Spikes: If a particular month shows a sudden increase, it could be due to marketing campaigns, discounts, or product launches. Similarly, a drop could suggest an off-season or external factors affecting demand.

2. Daily Sales Trends (Not shown but can be analyzed)
If plotted daily, sales might show weekend vs. weekday variations (e.g., more sales on Fridays and weekends).
Holidays and promotions could create short-term spikes on specific days.
Some dips may indicate low-demand periods or external disruptions.

3. Yearly Sales Trends (If more years were available)
A steady upward movement would suggest growth over multiple years.
Recurring high and low months across years indicate strong seasonal effects.
If sales drop year-over-year, it may signal market decline or external business challenges.

Conclusion
Sales are not constant and show fluctuations over months.
There are high and low points, which might indicate seasonal demand or business strategies.

# 3.Is the current pricing strategy effective?**bold text**

```python
pivot_table = df.pivot_table(values="Sales_Amount", index="Region", columns="Customer_Type", aggfunc="sum")

# Plot heatmap
plt.figure(figsize=(8,5))
sns.heatmap(pivot_table, cmap="coolwarm", annot=True, fmt=".0f")

plt.title("Sales Amount by Region and Customer Type")
plt.xlabel("Customer Type")
plt.ylabel("Region")
plt.show()
```

```python
df[["Quantity_Sold", "Unit_Price", "Sales_Amount"]].corr()
```

```python
plt.figure(figsize=(8, 5))
sns.regplot(data=df, x="Unit_Price", y="Quantity_Sold", scatter=False, color="red")
plt.xlabel("Unit Price")
plt.ylabel("Quantity Sold")
plt.title("Relationship Between Unit Price and Quantity Sold")
plt.show()
```

Negative Trend: The red regression line slopes downward, indicating an inverse relationship between price and quantity sold. 2.Lower-Priced Products Sell More: Many points are concentrated at low prices with high quantities, suggesting that cheaper products sell in larger volumes. Higher-Priced Products Sell Less: As Unit Price increases, Quantity Sold generally decreases, meaning expensive items have lower demand. Conclusion: Lower-priced products tend to sell more than higher-priced ones, which suggests price sensitivity among customers. To maximize revenue, the company could focus on volume-based pricing or introduce premium pricing strategies for high-value products

```python
df["Profit"] = df["Unit_Price"] - df["Unit_Cost"]
df.head()
plt.figure(figsize=(8, 5))
sns.regplot(data=df, x="Discount", y="Profit", scatter=False, color="red")
plt.xlabel("discount")
plt.ylabel("Quantity Sold")
plt.title("Relationship Between Unit Price and Quantity Sold")
plt.show()
```

Discounts help boost sales revenue but at the cost of reducing profit margins. Best Strategy: 1.Find an optimal discount percentage where sales increase without significantly harming profit. 2.Use targeted discounts (e.g., for slow-moving products or loyal customers). 3Analyze by product category to see where discounts are most effective.

```python
payment_counts = df["Payment_Method"].value_counts()
plt.figure(figsize=(8, 5))
sns.barplot(x=payment_counts.index, y=payment_counts.values, palette="coolwarm")
plt.xlabel("Payment Method")
plt.ylabel("Number of Transactions")
plt.title("Comparison of Payment Method Usage")
plt.xticks(rotation=45)
plt.show()
```
